```python
#Dagney Cooke Homework

# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Import data files
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the files into DataFrames
mouse_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# Merge dataframes based on Mouse ID
all_data = pd.merge(mouse_data, trial_data, on="Mouse ID")

# Display dataframe
all_data.head()
```

## Tumor Response to Treatment

```python
# Group data by Drug and Timepoint, and then calculate mean
grouped_means = all_data.groupby(["Drug","Timepoint"],as_index=False).mean()

# Save grouped data to dataframe
drug_timepoint = pd.DataFrame(grouped_means)

# remove Metastatic Sites column
drug_timepoint = drug_timepoint.drop(columns=["Metastatic Sites"])

# display data frame
drug_timepoint.head()
```

```python
# group data by Drug and Timepoint and then calculate standard error
tumor_vol_std = all_data.groupby(["Drug","Timepoint"],as_index=True).sem()

# convert data to list
tumor_vol_column = tumor_vol_std["Tumor Volume (mm3)"].tolist()

# copy the dataframe from above in preparation
drug_timepoint_stderror = drug_timepoint

# add std error column to dataframe
drug_timepoint_stderror["Tumor Volume"] = tumor_vol_column

# remove original column with means data
drug_timepoint_stderror = drug_timepoint_stderror.drop(columns=["Tumor Volume (mm3)"])

# change column title to Tumor Volume (mm3)
drug_timepoint_stderror = drug_timepoint_stderror.rename(columns={"Tumor Volume": "Tumor Volume (mm3)"})

# Display data frame
drug_timepoint_stderror.head()
```

```python
# Pivot data frame by moving Drug values from index to columns
drug_timepoint_rf = drug_timepoint.pivot(index="Timepoint", columns="Drug", values="Tumor Volume (mm3)")

# Display data frame
drug_timepoint_rf.head()
```

```python
# make list of drugs that we're looking for
drug_list = ["Capomulin","Infubinol","Ketapril","Placebo"]

# remove (trim) all drugs that we're not inerested in from both data frames
drug_timepoint_trimmed = drug_timepoint[drug_timepoint["Drug"].isin(drug_list)] 
drug_timepoint_stderror_trimmed = drug_timepoint_stderror[drug_timepoint_stderror["Drug"].isin(drug_list)] 

# set up figure
fig, ax = plt.subplots()

# create x axis list based on timepoints in column
x_axis = all_data["Timepoint"].unique()
x_axis = x_axis.tolist()

# set markers in advance
markers = ["o","^","s","D"]

# set colors in advance
colors = ["red","blue","green","black"]

# set labels to the drug names
labels = ["Capomulin","Infubinol","Ketapril","Placebo"]

# set up loop variables
i = 0
count = 0

# loop through data and add subplots to the figure
for i in range(0,len(drug_timepoint_trimmed["Tumor Volume (mm3)"]),10):
    ax.errorbar(x_axis, # x_axis is timepoints
                drug_timepoint_trimmed.iloc[i:i+10,2].tolist(), # y axis is tumor volume means
                yerr=drug_timepoint_stderror_trimmed.iloc[i:i+10,2].tolist(), # y axis is tumor volume std errors
                label=labels[count], # drug names
                fmt=markers[count], # marker style type
                color = colors[count], # color
                linestyle="--", # line style is dashed
                linewidth = .8, # adjust line width
                markersize = 5 # adjust marker size
                )
    count = count +1 # increase count for labels/markers/colors
    
plt.grid(axis='y') # set grid lines for y axis only

# set titles and labels
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# display legend
plt.legend()

# save figure
plt.savefig("Images/TumorResponseToTreatment.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Group data by drug and timepoint and calculate means
grouped_means = all_data.groupby(["Drug","Timepoint"],as_index=True).mean()

# Store data in dataframe
metastatic_means = pd.DataFrame(grouped_means)

# Remove Tumor Volume (mm3) column
metastatic_means = metastatic_means.drop(columns=["Tumor Volume (mm3)"])

# Display dataframe
metastatic_means.head()
```

```python
# Group data by drug and timepoint and calculate standard errors
grouped_stderr = all_data.groupby(["Drug","Timepoint"],as_index=True).sem()

# convert data to list (for later)
metastatic_stderr_list = grouped_stderr["Metastatic Sites"].tolist()

# create data frame
metastatic_stderr = pd.DataFrame(grouped_stderr)

# drop columns that don't matter
metastatic_stderr = metastatic_stderr.drop(columns=["Mouse ID","Tumor Volume (mm3)"])

# display dataframe
metastatic_stderr.head()
```

```python
# Group means with as_index=False (instead of index=True)
grouped_means_rf = all_data.groupby(["Drug","Timepoint"],as_index=False).mean()

# Create datafarme
metastatic_means_rf = pd.DataFrame(grouped_means_rf)

# Drop tumor volume column
metastatic_means_rf = metastatic_means_rf.drop(columns=["Tumor Volume (mm3)"])

# Move drug from index to column
metastatic_means_pivot = metastatic_means_rf.pivot(index="Timepoint", columns="Drug", values="Metastatic Sites")

# display dataframe
metastatic_means_pivot.head()
```

```python
# copy data frame
metastatic_total_data_rf = metastatic_means_rf

# add std error column to data frame
metastatic_total_data_rf["Std Error"] = metastatic_stderr_list

# display dataframe
metastatic_total_data_rf.head()
```

```python
# same list as earlier 
drug_list = ["Capomulin","Infubinol","Ketapril","Placebo"]

# remove drug info that's not in list
metastatic_data_trimmed = metastatic_total_data_rf[metastatic_total_data_rf["Drug"].isin(drug_list)] 

# set up figure
fig, ax = plt.subplots()

# get unique timepoints for x_axis and save in list
x_axis = all_data["Timepoint"].unique()
x_axis = x_axis.tolist()

# same markers/colors/labels as earlier
markers = ["o","^","s","D"]
colors = ["red","blue","green","black"]
labels = metastatic_data_trimmed["Drug"].unique().tolist()

# set up variables for loop
i = 0
count = 0

# loop through drugs and create subplots
for i in range(0,len(metastatic_data_trimmed["Metastatic Sites"]),10):
    ax.errorbar(x_axis, # x_axis is timepoints
                metastatic_data_trimmed.iloc[i:i+10,2].tolist(), # y axis is metastatic tumor means
                yerr=metastatic_data_trimmed.iloc[i:i+10,3].tolist(), # metastatic tumor std errors
                label=labels[count], # labels are drug names
                fmt=markers[count], # marker style
                color = colors[count], # marker color
                linestyle="--", # line style is dashed
                linewidth = .8,
                markersize = 5
                )
    count = count +1 # increase count for every drug to loop through markers/colors/labels
    
# show legened
plt.legend()

# adjust ylim to properly show data
plt.ylim(0, max(metastatic_data_trimmed["Metastatic Sites"])+.5)

# show y gridlines
plt.grid(axis = "y")

# label graph title and axes
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")

# save figure
plt.savefig("Images/MetastaticSpreadDuringTreatment.png")
# show graph
plt.show()
```

## Survival Rates

```python
# Group data by drug and timepoint and then get counts
grouped_counts = all_data.groupby(["Drug","Timepoint"],as_index=False).count()

# Convert to DataFrame
counts_data = pd.DataFrame(grouped_counts)

# Remove unnecessary columns
counts_data = counts_data.drop(columns=["Tumor Volume (mm3)","Metastatic Sites"])

# Preview DataFrame
counts_data.head()
```

```python
# Move drug from index to columns
counts_data_rf = counts_data.pivot(index="Timepoint", columns="Drug", values="Mouse ID")

# Preview the Data Frame
counts_data_rf.head(20)
```

```python
# get list of highest counts (ie original number of mice for every drug)
max_values = counts_data_rf.max().tolist()

# copy data frame
counts_max = counts_data

# add percent column to data frame
counts_max["Percent"] = np.arange(0,len(counts_max["Mouse ID"]))

# set up variables for loop
i = 0
j = 0

# loop through list and calculate percent mice alive at every row
for i in range(0,len(counts_max["Mouse ID"])):
    counts_max.iloc[i,3] = (counts_max.iloc[i,2]/max_values[j])*100 # calculate percent
    if (i%10) == 0: # for every 10 timepoints, move to the next original mice value
        j = j + 1 # increase mice count
        if j == 10: # once you get to 10 mice, break loop
            break

# display dataframe
counts_max.head()
```

```python
# same drug list
drug_list = ["Capomulin","Infubinol","Ketapril","Placebo"]

# remove unnecessary info from list
counts_max_trimmed = counts_max[counts_max["Drug"].isin(drug_list)] 

# set up figure
fig, ax = plt.subplots()

# get x axis timepoints
x_axis = all_data["Timepoint"].unique()
x_axis = x_axis.tolist()

# set up markers/colors/labels for graph
markers = ["o","^","s","D"]
colors = ["red","blue","green","black"]
labels = drug_list

# set up variables for loop
i = 0
count = 0

# loop through data and make subplots for each drug
for i in range(0,len(counts_max_trimmed["Percent"]),10):
    ax.plot(x_axis, # x axis is timepoints
                counts_max_trimmed.iloc[i:i+10,3].tolist(), # y axis is percent column values
                label=labels[count], # drug names
                c = colors[count], # colors
                marker = markers[count], # marker style
                linestyle="--", # dashed lines
                linewidth = .8
                )
    count = count +1 # move on to the next drug
    
# display legened
plt.legend()

# display grid
plt.grid()

# display titles/axes labels
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

# save figure
plt.savefig("Images/SurvivalDuringTreatment.png")
# show graph
plt.show()
```

## Summary Bar Graph

```python
# get mean data for timepoint 45 and save to list
size_45_data = drug_timepoint[drug_timepoint["Timepoint"] == 45]
size_45_list = size_45_data["Tumor Volume (mm3)"].tolist()

# get mean data for timepoint 0 and save to list
size_0_data = drug_timepoint[drug_timepoint["Timepoint"] == 0]
size_0_list = size_0_data["Tumor Volume (mm3)"].tolist()

# calculate percent change and store in list
percent_changes = [((x-y)/y)*100 for x,y in zip(size_45_list,size_0_list)]

# store total list of drugs in list
full_drug_list = drug_timepoint["Drug"].unique().tolist()
```

```python
# Store drug names and percent changes in tuple
drug_percent_change = [[full_drug_list[i],percent_changes[i]] for i in range(0,len(full_drug_list))]

# set up figure 
fig, ax = plt.subplots()

# set up variables for loop
i = 0
j = 0
count = 0
values = []

# loop through drug data
for i in range(0,len(drug_percent_change)):
    for j in range(0,len(drug_list)):
        if drug_percent_change[i][0] == drug_list[j]: # if the drug matches the drugs in the list, make graph
            if drug_percent_change[i][1] > 0: # if the drug change is negative, make it green and set height of text
                color_type = "red"
                height = 5
            elif drug_percent_change[i][1] < 0: # if the drug change is positive, make it red and set height of text
                color_type = "green"
                height = -5
            ax.bar(drug_list[j], # add bar to graph with x value as drug name
                   drug_percent_change[i][1], # drug percent change is height
                   color=color_type, # set color based on above if statement
                   width=1.0) # set width to remove any space between bars
            plt.text(count,height,str(round(drug_percent_change[i][1])) + "%",color='w') # set text label of % value
            count = count + 1 # increase count for text labels
            
# show grid
plt.grid()

# show titles and axis labels
plt.title("Tumor Change Over 45 Day Treatment")
plt.xlabel("Drug")
plt.ylabel("% Tumor Volume Change")

# create y axis tick labels
ytick_list = []
i = 0
for i in range(0,round(min(percent_changes)),-20):
    ytick_list.append(i)

for i in range(0,round(max(percent_changes)),20):
    ytick_list.append(i+20)

#ytick_list create based off of zero value
plt.yticks(ytick_list)
plt.margins(y = .1)

# save figure
plt.savefig("Images/TumorChangeOver45DayTreatment.png")

# show figure
plt.show()
```

Observable Trends

1. Survival rates with Infubinol and Ketapril are just as bad as if there was no drug at all (placebo).  Since mice don't know if they're being given drugs or not, this is a more realistic negative control than in human studies where placebos have been shown to have real medical effects.  
2. Capolumin both decreases the number of tumor sites and the sizes of the tumors.
3. Ramicane has a similar survival rate to Capolumin and a slightly lower average tumor size after 45 days of treatment.
